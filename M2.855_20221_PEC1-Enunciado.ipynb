En este ejercicio se aplicarán métodos de reducción de la dimensionalidad al conjunto original de datos. El objetivo es reducir el conjunto de atributos a un nuevo conjunto con menos dimensiones, pero que contengan la máxima información posible presente en los atributos originales.

Relizad las siguientes acciones:
- Aplicad el método de reducción de la dimensionalidad **Principal Component Analysis** (PCA) para reducir a 2 dimensiones el dataset original que contiene todos los atributos.

- Generad un **gráfico** con el resultado del PCA en el que se muestre, en función de los valores de las dos componentes en los ejes X e Y, el valor de la respuesta (_target_) usando la escala de colores. El objetivo es visualizar la variación del atributo objetivo en función de los valores de las componentes principales generadas.

**Nota**: 
- Vigilad de no incluir la variable objetivo en la reducción de dimensionalidad, dado que queremos poder explicar la variable objetivo en función del resto de variables reducidas a dos dimensiones.
- No es necesario que programéis el algoritmo de PCA, podéis usar la implementación disponible en la librería de `scikit-learn`.
- Podéis usar el _scatter plot_ de `matplotlib` con el parámetro `c`, que indica el color de los puntos, igual a la variable objetivo para generar el gráfico.

```python
# primero eliminamos de la matriz de datos la variable Y que es la que vamos a querer predecir:

cuantnorX = norm_df(cuant.drop(columns=['Y']))
diab_corrX = cuantnorX.corr()
```

```python
# una manera sencilla de reducir la dimensionalidad para obtener los dos componentes principales es utilizar la función de svd que genereará una PCA.
# Por el método de Singular Value Decomposition (SVD) obtenemos los dos primeros componentes principales.

U, s, Vt = np.linalg.svd(diab_corrX) # svd

pc2 = Vt.T[:, 0:2]
pc2
```

```python
df_pca = cuantnorX.dot(pc2)
df_pca['Y'] = cuantnor.Y
df_pca
```

```python
df_pca.info()
```

```python
sns.scatterplot(data=df_pca, x=0, y="Y",
                    color='red')
sns.scatterplot(data=df_pca, x=1, y="Y",
                    color='blue')
```

```python
sns.scatterplot(data=df_pca, x=0, y=1,
                    hue='Y')
```

Se puede ver que los valores de la variable Y correlacionan negativamente con las dos componentes principales. Esto es, cuanto más negativos son los valores de la 1a y 2a Componente Principal, mayor es la probabilidad de que la variable Y sea elebada. Esto es que a menor sean estos valores, mayores son las probabilidades de que la enfermedad progrese.

Relizad las siguientes operaciones:
- Repetid la reducción de dimensionalidad, pero en este caso usando **TSNE**. Podéis encontrar más información sobre este algoritmo en el link: https://distill.pub/2016/misread-tsne/
- Igual que antes, generad un gráfico con el resultado del TSNE en el que se muestre, en función de los valores de las 2 nuevas dimensiones en los ejes X e Y, el valor de la respuesta (_target_) usando una escala de colores.

**Nota**:
- No es necesario que programéis el algoritmo TSNE, podéis usar la implementación disponible en la librería de `scikit-learn`.
- A parte de especificar el número de componentes, probad a usar los parámetros `learning_rate` y `perplexity`.
- Podéis usar el _scatter plot_ de `matplotlib` con el parámetro `c`, que indica el color de los puntos, igual a la variable objetivo para generar el gráfico.

```python
def tsne_plot(data, p, lr, nr, nc, n=2):
    diab_tsne = TSNE(n_components=n, perplexity=p,
                learning_rate=lr, random_state=42).fit_transform(data)

    df_tsne = pd.DataFrame({'Component 1': diab_tsne[:, 0], 
                            'Component 2': diab_tsne[:, 1],
                            'Y': cuantnor.Y})

    x = sns.scatterplot(data=df_tsne, x='Component 1', y='Component 2',
                        hue='Y', ax=axs[nr,nc], legend=False)
    return x
```

Una vez tenemos la función que nos va a devlover el algoritmo de TSNE, lo podemos utilizar para probar diferentes valores de perplexity y de learnign rate, para que nos devuelva dos atributos diferenciados el uno del otro y reducir así la dimensionalidad.

```python
fig, axs = plt.subplots(4, 4, figsize=(25,25))

# valores bajos de learning rate y de perplexity
tsne_plot(cuantnorX, 3, 3, 0, 0)
# valores bajos de learning rate y intermedios de perplexity
tsne_plot(cuantnorX, 50, 3, 0, 1)
# valores bajos de learning rate y altos de perplexity
tsne_plot(cuantnorX, 90, 3, 0, 2)
# valores bajos de learning rate y muy altos de perplexity
tsne_plot(cuantnorX, 900, 3, 0, 3)

# valores intermedios de learning rate y bajos de perplexity
tsne_plot(cuantnorX, 3, 30, 1, 0)
# valores intermedios de learning rate y intermedios de perplexity
tsne_plot(cuantnorX, 50, 30, 1, 1)
# valores intermedios de learning rate y altos de perplexity
tsne_plot(cuantnorX, 90, 30, 1, 2)
# valores intermedios de learning rate y muy altos de perplexity
tsne_plot(cuantnorX, 900, 30, 1, 3)

# valores altos de learning rate y bajos de perplexity
tsne_plot(cuantnorX, 3, 90, 2, 0)
# valores altos de learning rate y intermedios de perplexity
tsne_plot(cuantnorX, 50, 90, 2, 1)
# valores altos de learning rate y altos de perplexity
tsne_plot(cuantnorX, 90, 90, 2, 2)
# valores altos de learning rate y muy altos de perplexity
tsne_plot(cuantnorX, 900, 90, 2, 3)

# valores muy altos de learning rate y bajos de perplexity
tsne_plot(cuantnorX, 3, 900, 3, 0)
# valores muy altos de learning rate y intermedios de perplexity
tsne_plot(cuantnorX, 50, 900, 3, 1)
# valores muy altos de learning rate y altos de perplexity
tsne_plot(cuantnorX, 90, 900, 3, 2)
# valores muy altos de learning rate y muy altos de perplexity
tsne_plot(cuantnorX, 900, 900, 3, 3)

#fig.suptitle('Perplexity', fontsize=12)

#fig.ylabel("Learning rate")
```

La métrica MAE es la más sencilla y penaliza de igual manera todos los errores independientemente de la grandaria del error. MAE mostrará que el error es pequeño si hay pocas instancias equivocadas. 

En cambio, MSE añade un aliciente. Tiene en cuenta la magnitud del error. De manera que los errores más grandes influirán más en la métrica que los errores pequeños. Esta distinción no estaba presente en MAE. Existe una modificación del MSE, y es el RMSE. El error de esta métrica se expresa en el mismo rango que el conjunto de datos. Lo que facilita la interpretación de los resultados. 

Por último, el MAPE devuelve el error como un porcentaje. Por lo que es más fácil de interpretar para un público más amplio (https://stephenallwright.com/rmse-vs-mape/).

Otro método útil para estimar la capacidad de predicción del modelo es utilizar una matriz de confusión como se veía en la teoría:

Para esto tendremos que calcular el valor predicho con el valor real de la variable Y

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(rfc, X_train, y_train, cv=5)

confusion_matrix(y_train, y_train_pred)
```

```python
tn, fp, fn, tp = confusion_matrix(y_train, y_train_pred).ravel()

prec = tp/(tp + fp)

sens = tp/(tp + fn)

spec = tn/(tn + fp)

print(f"La precisión del modelo es de {prec}, la sensibilidad de {sens} y la especifícidad de {spec}")
```

Podemos ver como los valores son muy bajos y el modelo tiene una capacidad muy baja para estimar el sexo de la persona con las variables utilizadas. Es por eso que sería necesario utilizar otros atributos para generar un modelo con mejorar capacidad de precisión.
